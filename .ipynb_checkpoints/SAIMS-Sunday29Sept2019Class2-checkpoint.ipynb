# Boolean Indexing


```python
data  = np.arange(64).reshape(8,8)
data
```

```python
bolarr = np.array([False, False, True, True, False, True, True, False],dtype="bool")
```

```python
bolarr = np.array(["Ali", "Nehal", "Hamza", "Umair", "Ali", "Nehal","Nasir","Nasir"])
cond = bolarr=="Ali"
cond
```

```python
data[cond]
```

```python
data
```

```python
data[[2,1,6,4,0]]
```

```python
cond  = (bolarr=="Ali") | (bolarr=="Nasir")
```

```python
data[~cond,3:6]
```

# Fancy Indexing

```python
arr = np.ones((8, 4),dtype = "int")
arr
```

```python
for i in range(8):
    arr[i]=i
arr
```

```python
arr[2]
```

```python
arr[[7,3,2,0],1:3]
```

```python
arr = np.arange(16).reshape(4,4)
```

```python
arr
```

```python
arr[[2,3,0,1],[2,1,0,3]]
#(2,2),(3,1),(0,0),(1,3)
```

```python
arr[[1, 0, 3, 2]] [0:2, [0, 3, 1, 2]]

# 1[0, 3, 1, 2], 0[0, 3, 1, 2],3[0, 3, 1, 2],2[0, 3, 1, 2]
# (1,0),(1,3),(1,1),(1,2)

```
